##Estimating the Size and Average Degree of Online Social Networks at the Extreme##
Abstract: Given the increasingly limiting nature of online
 social networks (OSNs), studying their structural characteristics
 under a limited data access model becomes important. In this
 study, we propose estimators for network size and average degree
 characteristics of OSNs. We sample an OSN graph using random
 neighbor API calls. Random neighbor API returns only the id
 of a randomly selected neighbor of a given user. Although the
 existing estimators give good accuracy estimations for a given
 sample size, they are not applicable under the extremely limited
 data access model considered here. We conduct experiments on
real world graphs to measure the performance of the proposed
 estimators.

### Print progress for AsyncResult to stdout ###

```python
def short_format_time(t):
    if t > 60:
        return "%4.1fmin" % (t / 60.)
    else:
        return " %5.1fs" % t
    
def wait_progress(ar, interval=5, timeout=-1):
    """Wait on an IPython AsyncResult, printing progress to stdout.
 
    Based on wait_interactive() in IPython and the output of Joblib in verbose mode.parallel
 
    This will work best when using a load-balanced view with a smallish chunk-size.
 
    """
    if timeout is None:
        timeout = -1
    N = len(ar)
    tic = time.time()
    print "\nRunning %i tasks:" % N
    sys.stdout.flush()
    last = 0
    while not ar.ready() and (timeout < 0 or time.time() - tic <= timeout):
        ar.wait(interval)
        progress, elapsed = ar.progress, ar.elapsed
        if progress > last:
            last = progress
            remaining = elapsed * (float(N) / progress - 1.)
            print '    Done %4i out of %4i | elapsed: %s remaining: %s' % (
                progress, N, short_format_time(elapsed), short_format_time(remaining))
            sys.stdout.flush()
    if ar.ready():
        try:
            speedup = round(100.0 * ar.serial_time / ar.wall_time)
            print "\nParallel speedup: %i%%" % speedup
        # For some reason ar.serial_time occasionally throws this exception. 
        # We choose to ignore it and just not display the speedup factor.
        except TypeError:
            pass
        
def get_graph(graph_file_list):
    for graph_file in graph_file_list:
        G=networkx.convert_node_labels_to_integers(networkx.connected_component_subgraphs(networkx.read_adjlist(graph_file))[0], first_label=0)
        G.name=graph_file.split('/')[-1][:-8]
        yield G
```

##1. Multiplicity based network size estimator ##
The main idea is to use vertex multiplcities instead of the (unknown) vertex degrees.

```python
%%px --local
def naive_network_size_estimator(S, multiplicity, margin=10):
    '''Estimate the network size using naive multiplicity approach
    
    Parameters
    ----------
    S            -- the sample
    multiplicity -- the number of times a vertex is selected (a function that returns the multiplicity given the vertex id)
    margin       -- the integer used to ignore the 'close' samples to reduce the dependency in the sample (useful in the random walk sampling)
    
    Note
    ----
    If degree is not known , vertex multiplicity is used instead in the estimation. If vertices are sampled independently, then margin should be 0; otherwise, we ignore the useful data.
    '''
    return estimate_N_NODE_Margin(S, weights=multiplicity, margin=margin)
```

###Monte carlo simulation for network size estimation###

```python
@lview.parallel(block=False, chunksize=1)
def mc_simulation_for_network_size_estimation(G, sampler, estimator, sample_size, margin_list, degree_known_list, number_of_simulations=1000):
    '''Perform a monte carlo simulation for network size estimation unsing naive algorithm (the algorithm using the vertex multiplicity insteda of vertex degree)
    
    Parameters
    ----------
    G                       -- population graph
    sampler                 -- sampler that returns a list of vertices from the population graph
    estimator               -- the estimator used to estimate the network size given the sample
    sample_size             -- size of the sample
    number_of_simulations   -- number of simulations (network size estimations) (default 1000)
    margin                  -- the positive value used to decrease the dependency of the samples (should be positive only if selections are not independent, e.g., random walk) (default 10)
    degree_known            -- boolean indicating whether the degree of a vertex is observable if sampled
    
    Return
    ------
    estimations                 -- a list of dictionaries that keep the parameters and the network size estimates
    '''
    
    estimations=[]
    for _ in range(number_of_simulations):
        S=[v for v in sampler(G, sample_size=int(sample_size))]
        for degree_known in degree_known_list:
            if degree_known:
                for margin in margin_list:
                    estimations.append({'graphName':G.name, 'networkSize': len(G),'averageDegree': 2.*G.number_of_edges()/len(G),'samplingFraction':1.*sample_size/G.number_of_nodes(),'margin':margin,\
                                        'degreeKnown':degree_known, 'networkSizeEstimate':estimator(S, G.degree, margin=margin)})
            else:
                for margin in margin_list:
                    estimations.append({'graphName':G.name, 'networkSize': len(G),'averageDegree': 2.*G.number_of_edges()/len(G),'samplingFraction':1.*sample_size/G.number_of_nodes(),'margin':margin,\
                                        'degreeKnown':degree_known, 'networkSizeEstimate':estimator(S, lambda x, mults=collections.Counter(S): mults[x], margin=margin)})

    return estimations
```

##3. Average degree estimator ##

###Monte carlo simulation for average degree estimation###

```python
%%px --local
def average_degree_estimator(S, degree_est, th=1):
    '''Estimate the network size using naive multiplicity approach
    
    Parameters
    ----------
    S            -- the sample
    th           -- the integer used to thin the sample data to reduce the dependency in the sample (useful in the random walk sampling)
    
    Note
    ----
    If degree is not known , vertex multiplicity is used instead in the estimation. If vertices are sampled independently, then th should be 0; otherwise, we ignore the useful data.
    '''
    return estimate_average_degree_ShiftedThinning(S, weights=degree_est, th=th)

def degree_estim(S,G,m):
    degree_dict={}
    for sample in S:
        C=0
        NS={}
        nbors=networkx.neighbors(G,sample)
        for _ in range(m):
            neigh=random.choice(nbors)
            
            try:
                NS[neigh]+=1
            except KeyError:
                NS[neigh] = 1

            C+=NS[neigh]-1
        try:
            degree_dict[sample] = (sum (degree_dict[sample]) +  ((m+1)*(m))/(2.*(C+1)) )/ (len(degree_dict[sample]) + 1)
        except:
            degree_dict[sample] = (m+1)*(m)/(2.*(C+1))
    return degree_dict
```

```python
@lview.parallel(block=False, chunksize=1)
def mc_simulation_for_average_degree(G, sampler, estimator, sample_size, m, thinning_list, degree_known_list, number_of_simulations=1000):
    '''Perform a monte carlo simulation for average degree estimation unsing def sampling
    
    Parameters
    ----------
    G                       -- population graph
    sampler                 -- sampler that returns a list of vertices from the population graph
    estimator               -- the estimator used to estimate the network size given the sample
    sample_size             -- size of the sample (total budget)
    m                       -- number of API call per vertex visit
    degree_known_list       -- boolean list indicating whether the degree of a vertex is observable (if sampled)
    number_of_simulations   -- number of simulations (network size estimations) (default 1000)
    Return
    ------
    estimations                 -- a list of dictionaries that keep the parameters and the network size estimates
    '''
    estimations=[]
    for _ in range(number_of_simulations):
        S=[v for v in sampler(G, sample_size=int(sample_size/m))]
        degree_dict=degree_estim(S, G, m)
        for degree_known in degree_known_list:
            for thinning in thinning_list:
                estimations.append({'graphName':G.name, 'networkSize': len(G),'averageDegree': 2.*G.number_of_edges()/len(G),'samplingFraction':1.*sample_size/G.number_of_nodes(),'m':m, 'degreeKnown':degree_known, \
                                    'avgDegreeEstimate': estimator(S, G.degree) if degree_known else estimator(S, lambda x, dic=degree_dict: dic[x],thinning), 'thinning':thinning})
            
    return estimations
```

```python
%%px --local
import collections
def mse(sim, obs):
    if type(sim) is not numpy.array:
        sim=numpy.array(sim)
    act=numpy.array(len(sim)*[obs])
    return numpy.mean(numpy.power(sim-act,2))

def nrmse(sim, obs, norm='relative'):
    rmse=numpy.sqrt(mse(sim,obs))
    result=0.0
    if norm is 'sd':
        result=rmse/std(sim)
    elif norm is 'maxmin':
        result=rmse/(max(sim)-min(sim))
    elif norm is 'mean':
        result=rmse/mean(sim)
    elif norm is 'relative' :
        result=rmse/obs
    elif norm is None:
        result=rmse
    else:
        print 'invalid norm'
    return result

def single_experiment_for_degree_estimation_for_fixed_m(m, degree):
    l = list((random.randint(0,degree) for _ in range(m)))
    freqs=collections.Counter(l)
    return sum((val*(val-1)/2. for key, val in freqs.iteritems()))

def mc_simulation_for_degree_estimate_for_fixed_m(m, degree, num_of_simulations=10000, onlyRes=True, corrected=True, ignoreInfEstimation=False):
    '''Perform monte carlo simulation for degree estimation
    
    Parameters
    ---------
    m                   -- number of API calls
    degree              -- degree of the vertex
    num_of_simulations  -- number of monte carlo simulations (default 10000)
    onlyRes             -- boolean to indicate whether ony the list of estimations will be returned (default True). returns a tuple (sampling fraction, res)
    corrected           -- whether the estimator is corrected to get smaller bias for small sample sizes, known as Chapman correction (default True)
    ignoreInfEstimation -- if correction is not made and C=0 estimation becomes infinity. If true, ignores the estimation. (default False)
    
    Return
    ------
    estimations                 -- a list of degree estimations if onlyRes is True, otherwise a tuple (sampling fraction, list of degree estimations)
    
    Note
    ----
    May throw exception if correction is not made, number of collusions is 0 and ignoreInfEstimation is False.
    '''
    estimations=[]
    for _ in range(num_of_simulations):
        C=single_experiment_for_degree_estimation_for_fixed_m(m, degree)
        if corrected:
            estimations.append((m+1)*(m)/(2.*(C+1))-1) 
        else: 
            if C==0 & ignoreInfEstimation:
                pass
            else:
                estimations.append((m-1)*(m)/(2.*C))
        
    return estimations if onlyRes else (1.*m/degree, estimations)
```

###Run the experiment for vertex degree estimation###

```python
def experiment_for_degree_estimation(m, degree, num_of_sims):
    return {'m':m, 'degree':degree, 'est': numpy.mean(mc_simulation_for_degree_estimate_for_fixed_m(m, degree, num_of_sims, onlyRes=True))/degree, 
                      'coeff_var': numpy.std(mc_simulation_for_degree_estimate_for_fixed_m(m, degree, num_of_sims, onlyRes=True))/numpy.mean(mc_simulation_for_degree_estimate_for_fixed_m(m, degree, num_of_sims, onlyRes=True)),
                      'nrmse':  nrmse(mc_simulation_for_degree_estimate_for_fixed_m(m, degree, num_of_sims, onlyRes=True), degree)    }

ms=[10,50,100, 200, 500,1000,2000]
degrees=[10,100,1000]
num_of_sims=[10000]
prod=zip(*list(itertools.product(ms,degrees, num_of_sims)))

l=[]
for result in lview.map(experiment_for_degree_estimation, *prod):
    l.append(result)
    print result

df=pandas.DataFrame(l)
pickle.dump(df,open('bar_plot_degree_estimation.pickle','w'))
```

### Sample size required for a given precision and accuracy###

```python
from scipy import stats
from numpy import arange

def plot_for_fixed_confidence_level(alpha=0.1, ax=None):
    epsilon=arange(0.1,1.1, 0.1)
    degrees=[1000,100,10]

    critical_value=stats.norm.ppf(1-alpha/2.)
    marker = itertools.cycle(('s','d','o','^')) 
    for degree in degrees:
        pandas.Series((1 + ( 1+ ( ( 8 *(degree-1)*critical_value**2 )/epsilon**2) )**0.5)/2., index=epsilon).plot(ax=ax, label=str(degree), marker=marker.next(), ms=8)
    ax.grid(False)
    ax.set_ylabel('Number of API calls ($m$)', fontsize=16)
    ax.set_xlabel(r'Relative error $(\epsilon)$',fontsize=16)
    ax.tick_params(axis='x',which='both',bottom='on',top='off', right='off',labelbottom='on')
    ax.tick_params(axis='y',which='both', right='off',labelbottom='on')
    ax.set_xlim(0.08,1.02)
    ax.set_ylim(0,1200)
    ax.set_xticks(epsilon)
    ax.text(0.05,0.94, '(a)', transform=ax.transAxes, ma= 'center', ha='left', va='top', fontsize=16)
    ax.text(0.15,0.94, '$90\%$ confidence level \n' + r'($\alpha=0.1$)', transform=ax.transAxes, ma= 'center', ha='left', va='top', fontsize=16, bbox=dict(facecolor='gray', alpha=0.1, edgecolor='black', boxstyle='round, pad=0.2'))
    #plt.legend(title='pop. degree')

fig=plt.figure(figsize=fig_dims)
ax=fig.gca()
plot_for_fixed_confidence_level(alpha=0.1, ax=ax)
#plt.savefig(open('sample_size_needed_precision.pdf','w'), format='pdf', bbox_inches='tight' )
```

```python
from scipy import stats
from numpy import arange

def plot_for_fixed_relative_error(epsilon, ax=None):
    alpha= numpy.array([0.4,0.3,0.2,0.1,0.05,0.01])
    degrees=[1000,100,10]
    critical_value=stats.norm.ppf(1-alpha/2.)
    marker = itertools.cycle(('s','d','o','^')) 
    
    for degree in degrees:
        pandas.Series((1 + ( 1+ ( ( 8 *(degree-1)*critical_value**2 )/epsilon**2) )**0.5)/2., index= 1-alpha).plot(ax=ax, label=str(degree), marker=marker.next(), ms=8)
    ax.grid(False)
    #ax.set_ylabel('Sample size needed')
    ax.set_xlabel(r'Confidence level $ (100(1-\alpha)\%)$',fontsize=16)
    ax.tick_params(axis='x',which='both',bottom='on',top='off', right='off',labelbottom='on')
    ax.tick_params(axis='y',which='both', right='off',labelbottom='on')
    ax.set_xlim(0.58,1.01)
    ax.set_xticks(1-alpha)
    ax.text(0.05,0.94, '(b)', transform=ax.transAxes, ma= 'center', ha='left', va='top', fontsize=16)
    ax.text(0.15,0.94, r'relative error ($\epsilon=0.1$)', transform=ax.transAxes, ma= 'center', ha='left', va='top', fontsize=16, bbox=dict(facecolor='gray', alpha=0.1, edgecolor='black', boxstyle='round, pad=0.2'))
    #plt.legend(title='pop. degree', loc=2)
    
fig=plt.figure(figsize=fig_dims)
ax=fig.gca()
epsilon=0.1
plot_for_fixed_relative_error(epsilon, ax=ax)
#plt.savefig(open('sample_size_needed_conflevel.pdf','w'), format='pdf', bbox_inches='tight' )
```
